```python
from transformers import AutoTokenizer, AutoModelForMaskedLM
import torch
from torchviz import make_dot

# Load model
tokenizer = AutoTokenizer.from_pretrained("Shushant/nepaliBERT")
model = AutoModelForMaskedLM.from_pretrained("Shushant/nepaliBERT")

# Create dummy input
dummy_input = torch.randint(0, tokenizer.vocab_size, (1, 10))

# Forward pass
output = model(dummy_input)

# Create visualization graph
dot = make_dot(output.logits, params=dict(model.named_parameters()))
dot.format = 'png'
dot.render("nepaliBERT_architecture")
```
